```python
for column in quan:
    descriptive.loc["Mean", column] = dataset[column].mean()
    descriptive.loc["Median", column] = dataset[column].median()
    descriptive.loc["Mode", column] = dataset[column].mode()[0]
    
    # following will handle the NaN/missing values
    descriptive.loc["Q1:25%", column] = dataset.describe()[column]["25%"]
    descriptive.loc["Q2:50%", column] = dataset.describe()[column]["50%"]
    descriptive.loc["Q3:75%", column] = dataset.describe()[column]["75%"]
    descriptive.loc["99%", column] = np.percentile(dataset[column], 99)
    descriptive.loc["Q4:100%", column] = dataset.describe()[column]["max"]

    # IQR
    descriptive.loc["IQR", column] = descriptive[column]["Q3:75%"] - descriptive[column]["Q1:25%"]
    descriptive.loc["1.5_Rule", column] = 1.5 * descriptive[column]["IQR"]
    descriptive.loc["Lower_Boundary", column] = descriptive[column]["Q1:25%"] - descriptive[column]["1.5_Rule"]
    descriptive.loc["Upper_Boundary", column] = descriptive[column]["Q3:75%"] + descriptive[column]["1.5_Rule"]
    descriptive.loc["Min", column] = dataset.describe()[column]["min"]
    descriptive.loc["Max", column] = dataset.describe()[column]["max"]
    descriptive.loc["Has_Lower_Outlier", column] = dataset[column].min() < descriptive.loc["Lower_Boundary", column]
    descriptive.loc["Has_Upper_Outlier", column] = dataset[column].max() > descriptive.loc["Upper_Boundary", column]
```

```python
descriptive
```

```python
def has_outliers(quan):
    lower = []
    upper = []
    
    for column in quan:
    	if (dataset[column] < descriptive.loc["Lower_Boundary", column]).any():
    		lower.append(column)
    	if (dataset[column] > descriptive.loc["Upper_Boundary", column]).any():
    		upper.append(column)

    print("Lower outlier list", lower)
    print("Upper outlier list", lower)
    
    return len(lower) > 0 or len(upper) > 0
```

```python
# Check the dataset has outliers, BEFORE replacing the lower/upper boundary values
result = has_outliers(quan)
print("Has outliers in the dataset" if result else "Doesn't have dataset")
```

### Replacing Outliers

```python
for column in quan:
    lower = descriptive.loc["Lower_Boundary", column]
    upper = descriptive.loc["Upper_Boundary", column]

    # reset the lower boundary value if any values lesser outlair (it will update the original dataset)
    dataset.loc[dataset[column] < lower, column] = lower

    # reset the upper boundary value if any values greater outlair (it will update the original dataset)
    dataset.loc[dataset[column] > upper, column] = upper
```

```python
# create a function
def get_freq_table(column, dataset):
    freq_table = pd.DataFrame(
        columns=["Unique_Values", "Frequency", "Relative_Frequency", "Cumulative_Relative_Frequency"],
    )
    
    freq_table["Unique_Values"]=dataset[column].value_counts().index
    freq_table["Frequency"]=dataset[column].value_counts().values
    freq_table["Relative_Frequency"]=freq_table["Frequency"]/len(dataset)
    freq_table["Cumulative_Relative_Frequency"]=freq_table["Relative_Frequency"].cumsum()
    
    return freq_table
```

```python
get_freq_table("hsc_p", dataset)
```
