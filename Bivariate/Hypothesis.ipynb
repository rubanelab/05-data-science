```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv("PrePlacement.csv")
```

```python
from scipy.stats import ttest_ind
# N/A values accepted for t-test. So remove before the process
dataset = dataset.dropna()

male = dataset[dataset["gender"] == "M"]["salary"]
female = dataset[dataset["gender"] == "F"]["salary"]

# print(male)
ttest_ind(male, female)

# The p-value is 0.03 whichis less the 0.05 so we can accept
```

```python

# Check the similarity between Central Board and Others groups related to salary
# N/A values accepted for t-test. So remove before the process
dataset = dataset.dropna()

central = dataset[dataset["ssc_b"] == "Central"]["salary"]
others = dataset[dataset["ssc_b"] == "Others"]["salary"]

ttest_ind(central, others)

# The p-value is 0.53 whichis greater the 0.05 so we cannot accept
```

### Dependenat Samples = Paired T-Test
- Same group (Male) but different conditions (ssc_p, hsc-)
- Group must be categorical data and condition must be numerical data


```python
from scipy.stats import ttest_rel
dataset = dataset.dropna()

male_ssc = dataset[dataset["gender"] == "M"]["ssc_p"]
male_hsc = dataset[dataset["gender"] == "M"]["hsc_p"]

ttest_rel(male_ssc, male_hsc)
```

```python
# Assignment - 1 (Independent) Different groups with same target
# HSC_S [Commerce, Science] vs [MBA_P, MBA_P]

dataset = dataset.dropna()

commerce_etest = dataset[dataset["hsc_s"] == "Commerce"]["mba_p"]
science_mba = dataset[dataset["hsc_s"] == "Science"]["mba_p"]

ttest_ind(commerce_etest, science_mba)
# The p-value is 0.46 (46%), and the two groups doesn't have any similarities.
```

```python
# Assignment - 2 (Dependent) Same group with different targets
# HSC_S [Commerce] vs [MBA_P, ETEST_P]

dataset = dataset.dropna()

commerce_etest = dataset[dataset["hsc_s"] == "Commerce"]["etest_p"]
commerce_mba = dataset[dataset["hsc_s"] == "Commerce"]["mba_p"]

ttest_rel(commerce_etest, commerce_mba)
```

```python
# Since the p-value is 2.46 (246%), there are NO similarities for Commerce subject with etest and mba test
```

### Hypothesis Testing

```python
# Statement: Test whether SSC_P of Male and HSC_P of Male is same?
# Same means talking about similarities
# If significance level not given, assume 5% (0.05) will be default.

# Null Hypostheis = There is NO similarity between 10th marks and 12th pass mark of male.
```

```python
dataset = dataset.dropna()

male_ssc = dataset[dataset["gender"] == "M"]["ssc_p"]
male_hsc = dataset[dataset["gender"] == "M"]["hsc_p"]

ttest_rel(male_ssc, male_hsc)
```

```python
# If the p-value is less than 0.05 (5%) we can reject the H0 (Null hypothesis)
# As the result of this testing the p-value is 0.49 (49%), we CANNOT reject the Null Hypothesis
# Means ==> We accept the Null Hypothesis

# Result => There is NO similarity between 10th marks and 12th pass mark of male.
```
