### Percentile
- NumPy is a library to handle numerical calulations
- np.percentile(dataset, percentage_val)

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("Placement.csv")
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset["hsc_s"].value_counts()
```

```python
dataset.info()
```

```python
dataset.describe()
# It will show only for numerical data (Since its applicable for numbers)
```

```python
np.percentile(dataset["ssc_p"], 10)
# 10% of the data falls under the 52 marks
```

```python
np.percentile(dataset["ssc_p"], 15)
# 15% of the data falls under the 55 marks
```

```python
np.percentile(dataset["ssc_p"], 50)
# 50% of the data falls under the 67 marks, which is also same as median (mid point) value
```

```python
np.percentile(dataset["ssc_p"], 75)
# 75% of the data falls under the 75.7 marks
```

```python
# JUST CHECKING FOR 150% (of course data out of range)
# np.percentile(dataset["ssc_p"], 150)
# ValueError: Percentiles must be in the range [0, 100]
```

```python
val_99 = np.percentile(dataset["ssc_p"], 99)
val_100 = np.percentile(dataset["ssc_p"], 100)

print("Value at 99% is {}, and at 100% is {}".format(val_99, val_100))
```

```python
from Univariate import Univariate 

qual, quan = Univariate.get_qual_quan(dataset)
```

```python
descriptive = pd.DataFrame(
    index=["Mean", "Median", "Mode", "Q1:25%", "Q2:50%", "Q3:75%", "99%","Q4:100%"],
    columns=quan
)
```

```python
for column in quan:
    descriptive[column]["Mean"] = dataset[column].mean()
    descriptive[column]["Median"] = dataset[column].median()
    descriptive[column]["Mode"] = dataset[column].mode()[0]
```

```python
for column in quan:
    descriptive[column]["Mean"] = dataset[column].mean()
    descriptive[column]["Median"] = dataset[column].median()
    descriptive[column]["Mode"] = dataset[column].mode()[0]
    descriptive[column]["Q1:25%"] = np.percentile(dataset[column], 25)
```

```python
for column in quan:
    descriptive[column]["Mean"] = dataset[column].mean()
    descriptive[column]["Median"] = dataset[column].median()
    descriptive[column]["Mode"] = dataset[column].mode()[0]
    
    # descriptive[column]["Q1:25%"] = np.percentile(dataset[column], 25)
    # descriptive[column]["Q2:50%"] = np.percentile(dataset[column], 50)
    # descriptive[column]["Q3:75%"] = np.percentile(dataset[column], 75)
    # descriptive[column]["Q3:100%"] = np.percentile(dataset[column], 100)
    
    # following will handle the NaN/missing values
    descriptive[column]["Q1:25%"] = dataset.describe()[column]["25%"]
    descriptive[column]["Q2:50%"] = dataset.describe()[column]["50%"]
    descriptive[column]["Q3:75%"] = dataset.describe()[column]["75%"]
    descriptive[column]["99%"] = np.percentile(dataset[column], 99)
    descriptive[column]["Q4:100%"] = dataset.describe()[column]["max"]
```
